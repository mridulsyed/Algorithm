The prolem we are trying to solve is based on GAIT. *Gait is a person's manner of walking*. We can analyse gait using pose estimation and predict a vast domain of medical interest with computer vision based gait analysis. But, pose estimation from videos in 2D space limits the possiblities of features we can consider. You will realise the situation better when you watch the following videos. Have a look.
